**Q3: A few students from a university have taken three exams**

Name = Sam, Jacy, Tom, Steve

Python = 25, 26, 29, 28

Statistics = 23, 21, 19, 25

SQL = 29, 27, 28, 25

**Please write a program which calculates mean values (no in-built function for mean) of all three tests and print the highest mean value. Addition and other functions are allowed. Please also report who scored the highest mark in python using programming construct**

```python
# Define the lists
name = ['Sam', 'Jacy', 'Tom', 'Steve']
python = [25, 26, 29, 28]
statistics = [23, 21, 19, 25]
sql = [29, 27, 28, 25]

# Define the function to calculate mean
def mean(list):
    sum = 0
    for i in list:
        sum += i
    return sum/len(list)

# Calculate the mean values
python_mean = mean(python)
statistics_mean = mean(statistics)
sql_mean = mean(sql)

# Print the highest mean value
print('The highest mean value is: ', max(python_mean, statistics_mean, sql_mean))

# Print the name of the student who scored the highest mark in python
print('The student who scored the highest mark in python is: ', name[python.index(max(python))])
```

**Q4: You are working in a medical store. A patient came to your medical store and asked to buy 2 strips of paracetamol, 3 strips of azithromycin and 5 strips of Vitamin C. One strip of paracetamol costs Rs 35, one strip of azithromycin costs Rs 49 and one strip of vitamin c costs Rs. 33. Patient gave you Rs 2000. Please tell us what is the total cost of each medicine, total cost of all medicine and how much money you refunded to the patient?**

```python
paracetamol=35
azithromycin=49
vitamin_c=33

paracetamol_strip=2
azithromycin_strip=3
vitamin_c_strip=5

total_cost_paracetamol=paracetamol*paracetamol_strip
total_cost_azithromycin=azithromycin*azithromycin_strip
total_cost_vitamin_c=vitamin_c*vitamin_c_strip

total_cost_all_medicine=total_cost_paracetamol+total_cost_azithromycin+total_cost_vitamin_c

money_refunded=2000-total_cost_all_medicine

print("total cost of paracetamol is",total_cost_paracetamol)
print("total cost of azithromycin is",total_cost_azithromycin)
print("total cost of vitamin c is",total_cost_vitamin_c)
print("total cost of all medicine is",total_cost_all_medicine)
print("money refunded to the patient is",money_refunded)
```

**Q5: Accept a sentence as input and find the number of vowels in it. Assume that the sentence has no punctuation marks. For example: I am learning python contains 6 vowels. This function should be applicable for all other different sentences.**

```python
def vowel_count(sentence):
    vowels = "aeiou"
    count = 0
    for i in sentence:
        if i in vowels:
            count += 1
    return count

print(vowel_count("I am learning python"))
```
